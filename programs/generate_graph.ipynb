```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 1024
client_nums = [4, 8, 16, 32]
time_spent = []
goon_clients = []

for client_n in client_nums:
    try:
        with open(f"./../experiments/matmul/{N}_{client_n}_100.txt") as f:
            lines = f.readlines()
            time_spent.append(float(lines[0]))
            goon_clients.append(client_n)
    except:
        print(client_n)
```

```python
plt.plot(goon_clients, time_spent)
plt.title("Runtime vs Number of Shards \n Matrix Multiplication")
plt.xlabel("Number of Shards")
plt.ylabel("Runtime (s)")
```
